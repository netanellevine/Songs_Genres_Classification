```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
songs_df = pd.read_csv("data/songs_with_genre.csv", index_col=0, header=[0, 1, 2])
```

```python
songs_df.head()
```

```python
generes = songs_df["genre_id"]
genres_index = pd.read_csv("data/genres_index.csv", index_col=0)
```

```python
zero_songs = generes.loc[generes['genre_id'] == 27]
len(zero_songs)
```

```python
generes.columns = ["genre_id"]
```

```python
plt.hist(generes['genre_id'],range(0,163))
```

```python
genres_index
```

```python
existsing = []
non_existing = []
for genre_id in genres_index['genre_index']:
    curr_len = len(generes.loc[generes['genre_id'] == genre_id])
    print(f"current genre {genre_id} has {curr_len} songs")
    if curr_len != 0:
        print()
        existsing.append(genre_id)
    else:
        non_existing.append(genre_id)
```

```python
existsing
```

```python
len(existsing)
```

```python
genres_no_zeros = genres_index.copy()
```

```python
non_existing
```

```python
for genre in genres_no_zeros.index:
    if genres_no_zeros.loc[genre][0] in non_existing:
        genres_no_zeros.drop([genre],inplace=True)
```

```python
genres_no_zeros
```

```python
new_range = range(len(genres_no_zeros['genre_index']))
```

```python
genres_no_zeros['genre_index'] = new_range
```

```python
genres_no_zeros
```

```python
genres_no_zeros.to_csv("data/genres_no_zeroes.csv")
```
