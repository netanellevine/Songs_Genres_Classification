```python
updated_songs = pd.read_csv("data/songs_with_genre.csv", index_col=0,header=[0, 1, 2])
updated_songs.head()
```

```python
updated_songs.rename(columns={'Unnamed: 519_level_1':'genre_id','Unnamed: 519_level_2':'genre_id'},inplace=True)
```

```python
updated_songs.head()
```

```python
out_csv = updated_songs.copy()
out_csv = out_csv.iloc[0:0]
```

```python
out_csv.head()
```

```python
generes = updated_songs['genre_id']['genre_id']['genre_id'].unique()
generes
```

```python
generes_dict_out = {genre:0 for genre in generes}
```

```python
generes_dict_out
```

```python
count = 0
for index,row in updated_songs.iterrows():
    genre = row['genre_id']['genre_id']['genre_id']
    length = len(updated_songs.loc[updated_songs['genre_id']['genre_id']['genre_id'] == genre])
    if length > 6000:
        count += 1
        generes_dict_out[genre] = generes_dict_out[genre]+1 
        out_csv = out_csv.append(row)
        updated_songs.drop(index,inplace=True)
    if count%1000 == 0:
        print(f"moved {count} songs")
```
